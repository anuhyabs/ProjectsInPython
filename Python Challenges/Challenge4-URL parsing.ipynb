```python
import urllib.request
from urllib.parse import *

url = "http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing=12345"
while(url!= None):
    print(url)
    request_url = urllib.request.urlopen(url)
    str = request_url.read()
    words = str.decode('utf-8').split()
    ll = words[-1]
    parse_url = urlparse(url)
    query = parse_url.query
    url_dict = dict(urllib.parse.parse_qs(query))
    url_dict['nothing'] = ll
    new_query = urllib.parse.urlencode(url_dict)
    parse_url = parse_url._replace(query=new_query)
    new_url = urllib.parse.urlunparse(parse_url)
    print(new_url)
    url=new_url
```

```python
import urllib, re
prefix = "http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing="
findnothing = re.compile(r"nothing is (\d+)").search
nothing = '12345'
while True:
    text = urllib.request.urlopen(prefix + nothing).read()
    print (text)
    match = findnothing(text)
    if match:
        nothing = match.group(1)
        print ("   going to", nothing)
    else:
        break
```
