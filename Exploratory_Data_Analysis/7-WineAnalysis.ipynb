    - alcohol

- Count the number of missing values in the entire dataset.

### Import Libraries

```python
import pandas as pd
```

### Load the data

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 
wine = pd.read_csv(url,header=None) 
```

### Exploring the data

```python
wine.head()
```

```python
# Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns.
wine = wine.drop(wine.columns[[0,3,6,8,10,12,13]],axis=1)
```

```python
# Assign different column names
col_names =['alcohol','malic_acid','alcalinity_of_ash','magnesium','flavanoids','proanthocyanins','hue']
wine.columns = col_names
```

```python
wine.head()
```

```python
# Set the values of the first 3 rows in the alcohol column as NaN
wine.iloc[0:3,0] = None
```

```python
wine.head()
```

```python
# Now set the value of the rows 3 and 4 of the magnesium column as NaN
wine.iloc[3:5,3] = None
```

```python
wine.head()
```

```python
# Fill in the null values (NaN) with the number 10 in the alcohol column and 100 in magnesium column.
wine['alcohol'] = wine.alcohol.fillna(10)
wine['magnesium'] = wine.magnesium.fillna(100)
```

```python
wine.head()
```

```python
# Count the number of missing values in the entire dataset.
wine.isna().sum()
```
